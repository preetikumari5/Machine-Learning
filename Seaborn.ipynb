Seaborn:


*   Data Visualisation Library




```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Note: Seaborn has some bulit-in datasets**

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for plots
sns.set_theme()
```

```python
#visualising data
sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
#house price
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
house = pd.DataFrame(housing.data,columns=housing.feature_names)
house['PRICE'] = housing.target
```

```python
house.head()
```

Distribution Plot

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
```

Correlation:


*   Positive
*   Negative



Heat Map

```python
cor = house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cor,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
