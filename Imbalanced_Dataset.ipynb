Imbalanced Dataset:


A dataset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/credit_data.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
#distribution of the two classes
data['Class'].value_counts()
```

0 --> legit transactions

1 --> fraudulent transactions

This is highly imbalanced dataset

```python
legit = data[data.Class == 0.0]
fraud = data[data.Class == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

Undersampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactions

```python
legit_sample = legit.sample(n=191)
```

```python
legit_sample.shape
```

Concatenate the Two DataFrames

```python
new_data = pd.concat([legit_sample , fraud] , axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```
