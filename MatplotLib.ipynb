MatplotLib:


*   Useful for making Plots








```python
# import lib
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine plot
plt.plot(x,z)
plt.show()
```

```python
#adding title,labels to axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine plot')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x**2')
plt.show()
```

```python
#two plots
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['java','python','php','javascript','c#','c++']
popularity = [22.2,17.6,8.8,8,7.7,6.7]
ax.bar(languages,popularity)
plt.xlabel('languages')
plt.ylabel('popularity')
plt.show()
```

### **Pie chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['java','python','php','javascript','c#','c++']
students = [50,25,23,86,46,100]
ax.pie(students,labels=langs,autopct='%1.1f%%')
plt.show()
```

### Scatter **Plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='y')
plt.show()
```

 **3D Scatter Plot**

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y= np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
